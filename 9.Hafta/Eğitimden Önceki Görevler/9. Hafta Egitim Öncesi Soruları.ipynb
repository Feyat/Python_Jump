# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
# 1.Problemin Çözümünü Buraya Yazınız
class Vehicle():
    def __init__(self, name, speed, mileage):
        self.name = name
        self.speed = speed
        self.mileage = mileage
    def intro(self):
        print(self.name, self.speed, self.mileage)

class Bus(Vehicle):
    def __init__(self, name, speed, mileage, capacity):
        super(). __init__(name, speed, mileage)
        self.capacity = capacity
    

    def seating_capacity(self):
        print(f'The seaating capacity of a bus is {self.capacity} passengers.')

a = Bus('Mercedes', 123, 234, 50)
a.seating_capacity()

# a = Bus('Mercedes', 180, 2453, 123)
#a.seating_capacity()
````
